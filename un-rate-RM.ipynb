```python

import matplotlib.pyplot as plt
import pandas as pd
yearly_df = pd.read_csv('UNRATE.csv', index_col=0)
plt.figure(figsize=(12, 6))


yearly_df['rolling_mean'] = yearly_df['UNRATE'].rolling(window=5).mean()
yearly_df['rolling_std'] = yearly_df['UNRATE'].rolling(window=5).std()


plt.plot(yearly_df.index, yearly_df['UNRATE'], label='Unemployment Rate', color='red', alpha=0.5)
plt.plot(yearly_df.index, yearly_df['rolling_mean'], label='5-Year Rolling Mean', color='black', linewidth=2)

plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.title('Rolling Mean of Inflation Rate')
plt.legend()
plt.grid(True)
plt.show()
```
